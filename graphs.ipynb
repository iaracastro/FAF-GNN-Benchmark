```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
random.seed(0)

# Small graph with two communities
G = nx.Graph()
G.add_nodes_from(range(10))

edges = [
    (0,1), (0,2), (1,2), (1,3), (2,3),
    (4,5), (4,6), (5,6), (5,7), (6,7),
    (3,4), (7,8), (8,9)
]
G.add_edges_from(edges)

G_rand = G.copy()
nx.double_edge_swap(
    G_rand,
    nswap=20,
    max_tries=200,
    seed=1
)

pos = nx.spring_layout(G, seed=2)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
nx.draw_networkx(
    G,
    pos=pos,
    node_color=["lightblue"]*4 + ["lightcoral"]*4 + ["lightgray"]*2,
    with_labels=True,
    node_size=600,
    font_size=9
)
plt.title("Original graph $G=(V,E,X,Y_L)$")
plt.axis("off")

plt.subplot(1, 2, 2)
nx.draw_networkx(
    G_rand,
    pos=pos,
    node_color=["lightblue"]*4 + ["lightcoral"]*4 + ["lightgray"]*2,
    with_labels=True,
    node_size=600,
    font_size=9
)
plt.title("Randomized graph $G_{rand}=(V,E_{rand},X,Y_L)$")
plt.axis("off")

#plt.suptitle("Same nodes, features, labels and splits; only topology is rewired")
plt.tight_layout()
plt.savefig("real_vs_random_graph.png", dpi=300, bbox_inches="tight")
plt.show()
```
